# **Q7:** `while loop` ka use karke user se tab tak numbers input lein jab tak wo `0` enter na karde. End mein saare numbers ka total dikhayein.

````python
total = 0

while True:
    num = int(input("Number enter karein (0 likhein band karne ke liye): "))
    if num == 0:
        break
    total = total + num

print("Saare numbers ka total:", total)
````

# **Q8:** Ek list `numbers = [12, 45, 7, 23, 56, 90, 11]` mein se sabse bada (Maximum) number find karein loop ke zariye.

````python
numbers = [12, 45, 7, 23, 56, 90, 11]

maximum = numbers[0]  

for num in numbers:
    if num > maximum:
        maximum = num

print("Sabse bada number:", maximum)
````

# **Q9:** User se ek string input lein (e.g., 'Python') aur loop ka use karke us string ko reverse print karein.

````python
text = input("Koi word likhein: ")

reversed_text = ""

for char in text:
    reversed_text = char + reversed_text

print("Ulta word:", reversed_text)
````

# **Q10:** 1 se 50 tak ke numbers check karein. Agar number 3 se divide hota hai to 'Fizz' print karein, agar 5 se hota hai to 'Buzz' print karein, aur agar dono se hota hai to 'FizzBuzz' print karein.

# **Q11:** Ek list di gayi hai `items = [1, 2, 3, 4, 5]`. Loop ka use karke ek nayi list banayein jis mein har number ka **Square** (e.g., 1, 4, 9...) ho.

````python
items = [1, 2, 3, 4, 5]
squares = []

for num in items:
    squares.append(num * num)

print("Squares ki list:", squares)
````

# **Q12:** User se ek number lein aur check karein ke wo **Prime Number** hai ya nahi (using loops).

````python
num = int(input("Koi number enter karein: "))

is_prime = True

if num < 2:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print(num, "ek Prime Number hai")
else:
    print(num, "Prime Number nahi hai")
````

# **Q15:** Ek list of lists di gayi hai: `matrix = [[1, 2], [3, 4], [5, 6]]`. Nested loop ka use karke saare individual numbers print karein.

# **Q16:** User se ek number input lein aur us ka **Factorial** find karein (e.g., 5! = 5*4*3*2*1).

````python
num = int(input("Koi number enter karein: "))

factorial = 1

for i in range(1, num + 1):
    factorial = factorial * i

print(num, "ka factorial:", factorial)
````

# **Q18:** Ek manual counting karein ke ek string (e.g., 'Learning Python is fun') mein kitne **Vowels** (a, e, i, o, u) hain.

````python
text = 'Learning Python is fun'
vowels = 'aeiouAEIOU'
count = 0

for char in text:
    if char in vowels:
        count = count + 1

print("Vowels ki tadaad:", count)
````

# **Q19:** **Fibonacci Sequence** ke pehle 10 numbers print karein (0, 1, 1, 2, 3, 5, 8...)

````python
a = 0
b = 1

for i in range(10):
    print(a)
    next_num = a + b
    a = b
    b = next_num
````

# **Q20:** Ek 'Number Guessing Game' banayein jahan program ek number (1-20) fix kare aur user ko 3 chances de guess karne ke liye. `while loop` aur `break` ka use karein.

````python
secret_number = 7  # yeh fixed number hai jo user ko guess karna hai
chances = 3

print("Number Guessing Game!")
print("1 se 20 ke beech ek number guess karein. Aapke paas 3 chances hain.")

while chances > 0:
    guess = int(input("Apna guess likhein: "))
    
    if guess == secret_number:
        print("Shahbash! Aapne sahi guess kiya!")
        break
    else:
        chances = chances - 1
        if chances > 0:
            print("Galat! Aapke paas", chances, "chance bacha hai.")
        else:
            print("Aapke saare chances khatam ho gaye. Sahi number tha:", secret_number)
````
