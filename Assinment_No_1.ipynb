### 43. Ek simple calculator banayein (+, -, *, /) symbols ke basis par operation perform karein.

```python
num1 = float(input("Enter first number: "))
operator = input("Enter operator (+, -, *, /): ")
num2 = float(input("Enter second number: "))

if operator == "+":
    print("Result:", num1 + num2)
elif operator == "-":
    print("Result:", num1 - num2)
elif operator == "*":
    print("Result:", num1 * num2)
elif operator == "/":
    if num2 != 0:
        print("Result:", num1 / num2)
    else:
        print("Error: Cannot divide by zero!")
else:
    print("Invalid operator!")
```

### 44. User ka weight aur height le kar BMI category batayein (Underweight, Normal, Overweight).

```python
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters (e.g. 1.75): "))

bmi = weight / (height ** 2)
print(f"Your BMI is: {bmi:.2f}")

if bmi < 18.5:
    print("Category: Underweight")
elif bmi < 25:
    print("Category: Normal")
else:
    print("Category: Overweight")
```

### 45. ATM Machine Logic: Pehle PIN check karein, agar sahi ho toh Amount poochiye, agar account balance se kam amount ho toh 'Withdrawal Success'.

```python
correct_pin = "1234"
account_balance = 5000

entered_pin = input("Enter your PIN: ")

if entered_pin == correct_pin:
    print(f"PIN correct! Your balance is: Rs {account_balance}")
    amount = float(input("Enter withdrawal amount: "))
    if amount <= account_balance:
        account_balance -= amount
        print("Withdrawal Successful!")
        print(f"Remaining balance: Rs {account_balance}")
    else:
        print("Insufficient balance!")
else:
    print("Wrong PIN! Access Denied.")
```

### 58. Electricity Bill: Pehle 100 units free, next 100 units Rs 5/unit, uske baad Rs 10/unit.

```python
units = int(input("Enter units consumed: "))

if units <= 100:
    bill = 0
elif units <= 200:
    bill = (units - 100) * 5
else:
    bill = (100 * 5) + (units - 200) * 10

print(f"Units consumed: {units}")
print(f"Total Electricity Bill: Rs {bill}")
```

### 60. Final Challenge: Ek mini-project logic banayein jisme user se Username, Password, aur ek Security Question poochein nested loops aur if-else ke saath.

```python
stored_username = "umair"
stored_password = "python123"
security_question = "What is your pet's name?"
security_answer = "fluffy"

print("=== Secure Login System ===")

username = input("Enter Username: ")

if username == stored_username:
    print("Username verified!")

    password = input("Enter Password: ")

    if password == stored_password:
        print("Password verified!")

        print("Security Question:", security_question)
        answer = input("Your answer: ").lower()

        if answer == security_answer:
            print("\n Login Successful! Welcome,", username + "!")
        else:
            print("\n Wrong answer to security question. Access Denied.")
    else:
        print("\n Wrong Password. Access Denied.")
else:
    print("\n Username not found. Access Denied.")
```
