**Q1:** Aik `Student` class banayein jismein `name` aur `roll_number` private variables (`__`) hon. In variables ko access aur modify karne ke liye getter aur setter methods likhein.

```python
class Student:
    def __init__(self, name, roll_number):
        self.__name = name
        self.__roll_number = roll_number

    # Getter methods
    def get_name(self):
        return self.__name

    def get_roll_number(self):
        return self.__roll_number

    # Setter methods
    def set_name(self, name):
        self.__name = name

    def set_roll_number(self, roll_number):
        self.__roll_number = roll_number


s = Student("Umair", 101)
print("Name:", s.get_name())
print("Roll Number:", s.get_roll_number())

s.set_name("Khan")
s.set_roll_number(202)
print("Updated Name:", s.get_name())
print("Updated Roll Number:", s.get_roll_number())
```

**Q2:** Aik `BankAccount` class banayein jismein `__balance` private variable ho. Aik `deposit()` aur aik `withdraw()` method banayein, check karein ke withdraw karte waqt balance zero se kam na ho.

```python
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited: {amount} | Balance: {self.__balance}")

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient balance!")
        else:
            self.__balance -= amount
            print(f"Withdrawn: {amount} | Balance: {self.__balance}")

    def get_balance(self):
        return self.__balance


acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(200)
acc.withdraw(2000)
```

**Q3:** Aik class banayein `Employee` jismein salary private ho. Agar koi salary ko minus (negative) value set karne ki koshish kare, toh setter method 'Invalid Salary' ka error message show kare.

```python
class Employee:
    def __init__(self, name, salary):
        self.__name = name
        self.set_salary(salary)

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        if salary < 0:
            print("Invalid Salary")
        else:
            self.__salary = salary

    def display(self):
        print(f"Employee: {self.__name}, Salary: {self.__salary}")


e = Employee("Sara", 50000)
e.display()

e.set_salary(-1000)  # Invalid case
e.set_salary(60000)  # Valid update
e.display()
```

**Q4:** Protected variable (`_`) aur Private variable (`__`) ke beech kya farq hota hai? Aik simple code ke zariye dono ka use-case dikhayein.

```python
class Person:
    def __init__(self):
        self._protected = "I am Protected"   # convention: don't access outside, but you can
        self.__private = "I am Private"       # strictly hidden (name mangling)

    def show(self):
        print(self._protected)
        print(self.__private)


class Child(Person):
    def access(self):
        print(self._protected)      # Accessible in child class
        # print(self.__private)     # This would cause AttributeError


p = Person()
p.show()

print(p._protected)        # Accessible (not recommended but works)
# print(p.__private)       # AttributeError

c = Child()
c.access()

# Private variable is accessible using name mangling (not recommended)
print(p._Person__private)
```

**Q5:** Encapsulation ka real-world faida kya hai? Private variable ko class ke bahar se direct access karne par kaunsa error aata hai? Code likh kar check karein.

```python
# Encapsulation ka faida:
# Data ko outside access/modification se protect karta hai.
# Sirf defined methods ke zariye data ko change kia ja sakta hai.
# Example: Bank ka balance directly change nahi hona chahiye.

class SecureData:
    def __init__(self):
        self.__secret = "Hidden Value"

    def get_secret(self):
        return self.__secret


obj = SecureData()
print(obj.get_secret())  # Works fine

try:
    print(obj.__secret)  # This will raise AttributeError
except AttributeError as e:
    print("Error:", e)
```

**Q6:** Single Inheritance ka use karte hue aik parent class `Vehicle` (with brand attribute) aur child class `Car` (with model attribute) banayein. Car ka object bana kar dono attributes print karein.

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand


class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

    def display(self):
        print(f"Brand: {self.brand}, Model: {self.model}")


c = Car("Toyota", "Corolla")
c.display()
```

**Q7:** Multilevel Inheritance ka use karein. Class `Animal` -> Class `Mammal` -> Class `Dog`. Dog class ke object se parent classes ke features ko access karein.

```python
class Animal:
    def breathe(self):
        print("Animals breathe oxygen.")


class Mammal(Animal):
    def feed_milk(self):
        print("Mammals feed milk to their young.")


class Dog(Mammal):
    def bark(self):
        print("Dog says: Woof!")


d = Dog()
d.breathe()      # From Animal
d.feed_milk()    # From Mammal
d.bark()         # Own method
```

**Q9:** Child class ke constructor (`__init__`) ke andar parent class ke constructor ko call karne ke liye `super()` ka use kaise hota hai? Aik practical example likhein.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"Person created: {self.name}, Age: {self.age}")


class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)   # Parent constructor call
        self.student_id = student_id
        print(f"Student ID: {self.student_id}")


s = Student("Usman", 20, "S-1001")
```

**Q10:** Hierarchical Inheritance kya hoti hai? Aik parent class `Shape` banayein aur us se do alag child classes `Circle` aur `Square` banakar samjhayein.

```python
class Shape:
    def describe(self):
        print("I am a shape.")


class Circle(Shape):
    def area(self, radius):
        print(f"Circle Area: {3.14 * radius * radius:.2f}")


class Square(Shape):
    def area(self, side):
        print(f"Square Area: {side * side}")


c = Circle()
c.describe()
c.area(5)

s = Square()
s.describe()
s.area(4)
```

**Q11:** Method Overriding kya hai? Parent class `Bird` banayein jisme `fly()` method ho jo print kare 'Most birds can fly'. Child class `Ostrich` banayein jo is method ko override kare aur print kare 'Ostriches cannot fly'.

```python
class Bird:
    def fly(self):
        print("Most birds can fly")


class Ostrich(Bird):
    def fly(self):
        print("Ostriches cannot fly")


b = Bird()
b.fly()

o = Ostrich()
o.fly()
```

**Q12:** Polymorphism function ke zariye kaise achieve hoti hai? Aik function banayein `make_sound(animal_object)` jo har tarah ke animal class ka sound method automatically call kare.

```python
class Dog:
    def sound(self):
        print("Dog says: Woof!")


class Cat:
    def sound(self):
        print("Cat says: Meow!")


class Cow:
    def sound(self):
        print("Cow says: Moo!")


def make_sound(animal_object):
    animal_object.sound()


animals = [Dog(), Cat(), Cow()]
for animal in animals:
    make_sound(animal)
```

**Q13:** Operator Overloading kya hoti hai? Do objects ko `+` operator ke zariye add karne ke liye `__add__` magic method ka use kaise karte hain? Code likhein.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def display(self):
        print(f"Point({self.x}, {self.y})")


p1 = Point(2, 3)
p2 = Point(4, 5)
p3 = p1 + p2   # __add__ called automatically
p3.display()
```

**Q17:** Q16 wali `Computer` class ko inherit karke do child classes `Laptop` aur `Desktop` banayein aur dono mein `process()` method ko mandatory implement karein.

```python
from abc import ABC, abstractmethod

class Computer(ABC):
    @abstractmethod
    def process(self):
        pass


class Laptop(Computer):
    def process(self):
        print("Laptop is processing on battery power.")


class Desktop(Computer):
    def process(self):
        print("Desktop is processing with full power.")


l = Laptop()
l.process()

d = Desktop()
d.process()
```

**Q18:** Aik abstract class `Car` banayein jisme aik concrete (normal) method `start_engine()` ho aur aik abstract method `drive()` ho. Child class bana kar dono ko test karein.

```python
from abc import ABC, abstractmethod

class Car(ABC):
    def start_engine(self):
        print("Engine started!")

    @abstractmethod
    def drive(self):
        pass


class ElectricCar(Car):
    def drive(self):
        print("Electric Car is driving silently.")


ec = ElectricCar()
ec.start_engine()   # Concrete method from abstract class
ec.drive()          # Overridden abstract method
```

**Q19:** Abstraction aur Encapsulation ke darmiyan kya wazeh farq (main difference) hai? Dono ke concepts ko short statements mei define karein.

```python
# Encapsulation:
# Data aur methods ko ek class mein band karna aur private/protected rakhna.
# Focus: Data ko HIDE karna aur protect karna.

# Abstraction:
# Complex details ko chupaana aur sirf zaroori cheezein show karna.
# Focus: Implementation HIDE karna, sirf interface dikhana.

# Simple Example:
# ATM machine use karte waqt aap sirf buttons dabaate ho (abstraction)
# Andar ki circuitry aur code chupi hoti hai (encapsulation)

print("Encapsulation = Data ko private rakhna (class ke andar band karna)")
print("Abstraction   = Sirf zaroori details dikhana, baqi hide karna")
```

**Q20:** Aik real-world scenario (jaise Payment System) design karein jahan `Payment` aik abstract class ho aur `CreditCardPayment` aur `UPIPayment` uski subclasses hon.

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    def __init__(self, amount):
        self.amount = amount

    @abstractmethod
    def pay(self):
        pass


class CreditCardPayment(Payment):
    def pay(self):
        print(f"Paid Rs.{self.amount} via Credit Card.")


class UPIPayment(Payment):
    def pay(self):
        print(f"Paid Rs.{self.amount} via UPI.")


payments = [CreditCardPayment(5000), UPIPayment(1500)]
for p in payments:
    p.pay()
```
